```python
# Main Clustering Analysis with Visual Evaluation (Aligned with Class Labs)

import data_preprocessor as dp

# -----------------------------
# Moons Dataset — Best for DBSCAN (Non-Spherical Clusters)
# -----------------------------

print("\n--- Moons Dataset Analysis ---")
X_moons = dp.generate_moons()

# DBSCAN on Moons
labels_dbscan_moons = dp.apply_dbscan(X_moons, eps=0.3, min_samples=5)
dp.plot_clusters_with_silhouette(X_moons, labels_dbscan_moons, 'DBSCAN on Moons')

# k-Means on Moons
labels_kmeans_moons = dp.apply_kmeans(X_moons, n_clusters=2)
dp.plot_clusters_with_silhouette(X_moons, labels_kmeans_moons, 'k-Means on Moons')

# Hierarchical Clustering on Moons
labels_hierarchical_moons = dp.apply_hierarchical(X_moons, n_clusters=2)
dp.plot_clusters_with_silhouette(X_moons, labels_hierarchical_moons, 'Hierarchical Clustering on Moons')

# Dendrogram for Moons
dp.plot_dendrogram(X_moons, method='ward', title='Dendrogram on Moons Dataset')

# -----------------------------
# Blobs Dataset — Shows DBSCAN Struggle with Varying Densities
# -----------------------------

print("\n--- Blobs Dataset Analysis ---")
X_blobs = dp.generate_blobs()

# DBSCAN on Blobs
labels_dbscan_blobs = dp.apply_dbscan(X_blobs, eps=0.5, min_samples=5)
dp.plot_clusters_with_silhouette(X_blobs, labels_dbscan_blobs, 'DBSCAN on Blobs')

# k-Means on Blobs
labels_kmeans_blobs = dp.apply_kmeans(X_blobs, n_clusters=3)
dp.plot_clusters_with_silhouette(X_blobs, labels_kmeans_blobs, 'k-Means on Blobs')

# Hierarchical Clustering on Blobs
labels_hierarchical_blobs = dp.apply_hierarchical(X_blobs, n_clusters=3)
dp.plot_clusters_with_silhouette(X_blobs, labels_hierarchical_blobs, 'Hierarchical Clustering on Blobs')

# Dendrogram for Blobs
dp.plot_dendrogram(X_blobs, method='ward', title='Dendrogram on Blobs Dataset')

```
