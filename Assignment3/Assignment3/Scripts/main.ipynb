```python
# Import necessary modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import data_preprocessor as dp

from sklearn.linear_model import ElasticNet, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, f1_score, roc_auc_score, average_precision_score, roc_curve, precision_recall_curve
from sklearn.model_selection import train_test_split
# 1. Load the dataset
messy_data = pd.read_csv('../Data/heart_disease_uci.csv')

# 2. Data Exploration Messy Data
print("Messy Data Info")
display(messy_data.info())
print("Messy Data Description")
display(messy_data.describe(include='all'))
print("Missing Values in Messy Data")
display(messy_data.isnull().sum())

# Data visualisation: Histograms
numeric_cols = messy_data.select_dtypes(include='number').columns
messy_data[numeric_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle("Messy Data - Histograms", fontsize=16)
plt.tight_layout()
plt.show()

# 3. Preprocess the data
print("Preprocess the data")
clean_data = messy_data.copy()
clean_data = dp.impute_missing_values(clean_data, strategy='mean')
clean_data = dp.remove_duplicates(clean_data)
clean_data = dp.normalize_data(clean_data)
clean_data = dp.remove_redundant_features(clean_data)

# 4. Save the cleaned dataset
clean_data.to_csv('../Data/clean_data.csv', index=False)

#5. Explore the cleaned dataset
print("Clean Data Info")
display(clean_data.info())
print("Clean Data Description")
display(clean_data.describe(include='all'))
print("Missing Values in Clean Data")
display(clean_data.isnull().sum())

# Histograms
numeric_cols = clean_data.select_dtypes(include='number').columns
clean_data[numeric_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle("Clean Data - Histograms", fontsize=16)
plt.tight_layout()
plt.show()

# 6. Simple Model Checks
print("\nSimple Model - Regression Task (ElasticNet on 'chol')")
dp.simple_model(clean_data, target_col='chol', scale_data=True)

print("\nSimple Model - Classification Task (Logistic Regression on 'num')")
clean_data['num'] = clean_data['num'].apply(lambda x: 1 if x > 0 else 0)
dp.simple_model(clean_data, target_col='num', scale_data=True, print_report=True)

# 7A. ElasticNet Regression Hyperparameter Tuning
alphas = [0.01, 0.1, 1, 10]
l1_ratios = [0.1, 0.5, 0.9]

regression_data = pd.get_dummies(clean_data.copy())
X_reg = regression_data.drop(columns=['chol'])
y_reg = regression_data['chol']

r2_results = pd.DataFrame(index=alphas, columns=l1_ratios)
rmse_results = pd.DataFrame(index=alphas, columns=l1_ratios)

for alpha in alphas:
    for l1_ratio in l1_ratios:
        X_train, X_test, y_train, y_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
        model = ElasticNet(alpha=alpha, l1_ratio=l1_ratio, random_state=42)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        r2_results.loc[alpha, l1_ratio] = r2_score(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        rmse_results.loc[alpha, l1_ratio] = np.sqrt(mse)

# Heatmaps
plt.figure(figsize=(10, 6))
sns.heatmap(r2_results.astype(float), annot=True, cmap='coolwarm')
plt.title('ElasticNet R Heatmap')
plt.xlabel('l1_ratio')
plt.ylabel('alpha')
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(rmse_results.astype(float), annot=True, cmap='coolwarm')
plt.title('ElasticNet RMSE Heatmap')
plt.xlabel('l1_ratio')
plt.ylabel('alpha')
plt.show()
       
# 7B. Logistic Regression Hyperparameter Tuning
penalties = ['l1', 'l2']
solvers = ['liblinear', 'saga']

X_class = pd.get_dummies(clean_data.drop(columns=['num']))
y_class = clean_data['num']

for penalty in penalties:
    for solver in solvers:
        try:
            print(f"\nLogistic Regression with penalty={penalty}, solver={solver}")
            X_train, X_test, y_train, y_test = train_test_split(X_class, y_class, stratify=y_class, test_size=0.2, random_state=42)
            model = LogisticRegression(penalty=penalty, solver=solver, max_iter=1000, random_state=42)
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
            print(f"F1 Score: {f1_score(y_test, y_pred):.4f}")
        except Exception as e:
            print(f"Skipped {penalty} + {solver} due to error: {e}")

# 7C. k-NN Classification with varying k
for k in [1, 5, 10]:
    print(f"\nK-Nearest Neighbors with n_neighbors={k}")
    from sklearn.neighbors import KNeighborsClassifier
    X_train, X_test, y_train, y_test = train_test_split(X_class, y_class, stratify=y_class, test_size=0.2, random_state=42)
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(f"F1 Score: {f1_score(y_test, y_pred):.4f}")

# 8. ROC & PR for Best Model

# Logistic Regression (l2, liblinear)
best_log_model = LogisticRegression(penalty='l2', solver='liblinear', max_iter=1000, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_class, y_class, stratify=y_class, test_size=0.2, random_state=42)
best_log_model.fit(X_train, y_train)
dp.plot_roc_pr(best_log_model, X_test, y_test, model_name='Logistic Regression (Best)')

# k-NN (n_neighbors=5)
best_knn_model = KNeighborsClassifier(n_neighbors=5)
X_train, X_test, y_train, y_test = train_test_split(X_class, y_class, stratify=y_class, test_size=0.2, random_state=42)
best_knn_model.fit(X_train, y_train)
dp.plot_roc_pr(best_knn_model, X_test, y_test, model_name='k-NN (Best, k=5)')

```
